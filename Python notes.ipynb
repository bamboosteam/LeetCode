```python
b = '1'
c = ''
```

```python
while b:
    temp = b[0]
    res = 0
    while temp == b[res]:
        res += 1
        if res + 1 > len(b):
            break
    b = b.lstrip(b[0])
    c += str(res)
    c += temp
    print('temp:{}, res:{}, c:{}'.format(temp, res, c))
```
